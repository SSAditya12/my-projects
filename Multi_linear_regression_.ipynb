```python
import kagglehub
import pandas as pd

# Download latest version
path = kagglehub.dataset_download("asinow/car-price-dataset")

print("Path to dataset files:", path)

# Construct the full file path
file_path = path + "/car_price_dataset.csv"

# Read the CSV using the constructed file path
b = pd.read_csv(file_path)
```

```python
b.head()
```

```python
b['Brand'] = b['Brand'].astype('category')
b['Brand'] = b['Brand'].cat.codes
b['Model'] = b['Model'].astype('category')
b['Model'] = b['Model'].cat.codes
b['Transmission'] = b['Transmission'].astype('category')
b['Transmission'] = b['Transmission'].cat.codes
b['Fuel_Type'] = b['Fuel_Type'].astype('category')
b['Fuel_Type'] = b['Fuel_Type'].cat.codes
```

```python
#x = b.iloc[:, : -1]
x = b.drop(columns = "Price")
```

```python
x
```

```python
#y = b.iloc[:,-1]
y = b["Price"]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.3,random_state=20)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
# Evaluation metrics
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(x_test.iloc[:, 0], y_test, label="Actual", alpha=0.5)  # Example: plotting against the first feature
plt.scatter(x_test.iloc[:, 0], y_pred, label="Predicted", alpha=0.5) # Example: plotting against the first feature
plt.xlabel(x_test.columns[0]) # Label the x-axis with the name of the first feature
plt.ylabel("Price")
plt.title("Actual vs. Predicted Prices (Example with First Feature)")
plt.legend()
plt.show()
```

```python
sorted_indices = np.argsort(x_test.iloc[:, 0])  # Sort by the first feature (example)
plt.figure(figsize=(8, 6))
plt.scatter(x_test.iloc[sorted_indices, 0], y_test.iloc[sorted_indices], label="Actual", alpha=0.5)
plt.plot(x_test.iloc[sorted_indices, 0], y_pred[sorted_indices], color='red', label="Predicted Trend")
plt.xlabel(x_test.columns[0]) # Label the x-axis with the name of the first feature
plt.ylabel("Price")
plt.title("Actual vs. Predicted Prices (Trend)")
plt.legend()
plt.show()
```

```python
# Plotting residuals (to check for patterns)
residuals = y_test - y_pred
plt.figure(figsize=(8, 6))
plt.scatter(y_pred, residuals, alpha=0.5)
plt.xlabel("Predicted Prices")
plt.ylabel("Residuals")
plt.title("Residual Plot")
plt.axhline(y=0, color='r', linestyle='--')  # Add a horizontal line at y=0
plt.show()
```

```python
#For plotting against all features (not easily visualizable, but you can iterate):
for i in range(x_test.shape[1]):  # Iterate through all features
    plt.figure(figsize=(8, 6))
    plt.scatter(x_test.iloc[:, i], y_test, label="Actual", alpha=0.5)
    plt.scatter(x_test.iloc[:, i], y_pred, label="Predicted", alpha=0.5)
    plt.xlabel(x_test.columns[i])
    plt.ylabel("Price")
    plt.title(f"Actual vs. Predicted Prices (Feature: {x_test.columns[i]})")
    plt.legend()
    plt.show()
```

```python
import pandas as pd
new_input = {
    'Brand' : ['Swift'],
    'Model' : ['Rio'],
    'Year' : ['2022'],
    'Engine_Size' : ['6'],
    'Fuel_Type': ['Petrol'],
    'Transmission': ['Automatic'],
    'Mileage': ['543256'],
    'Doors': ['4'],
   'Owner_Count': ['6'],

}
s = pd.DataFrame(new_input)


```

```python
# Apply categorical encoding to the new input data
# Assuming 'b' is the original DataFrame used for training
for column in ['Brand', 'Model', 'Fuel_Type', 'Transmission']:
    # Create a mapping of categories to codes from the original DataFrame 'b'
    categories = b[column].astype('category').cat.categories
    codes = b[column].astype('category').cat.codes

    # Create a dictionary for mapping categories to codes
    category_mapping = dict(zip(categories, codes))

    # Apply the mapping to the new input data
    s[column] = s[column].map(category_mapping)

# Convert all columns to numeric, handling errors by setting invalid values to NaN
for col in s.columns:
    s[col] = pd.to_numeric(s[col], errors='coerce')

# Fill NaN values with 0 (or another appropriate value)
s = s.fillna(0)

# Now you can make predictions
predictions = loaded_model.predict(s)
print(predictions)
```
