```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a =pd.read_csv("deliverytime.csv")
```

```python
a.head()
```

```python
a.info()
```

```python
a.describe()
```

```python
a['ID'] = a['ID'].astype('category')
a['ID']=a['ID'].cat.codes
a['Delivery_person_ID']=a['Delivery_person_ID'].astype('category')
a['Delivery_person_ID']=a['Delivery_person_ID'].cat.codes
a['Type_of_order']=a['Type_of_order'].astype('category')
a['Type_of_order']=a['Type_of_order'].cat.codes
a['Type_of_vehicle'] = a['Type_of_vehicle'].astype('category')
a['Type_of_vehicle']=a['Type_of_vehicle'].cat.codes
a.head()
```

```python
a.isnull().sum()
```

```python
a.dropna(inplace=True)
```

```python
a.isnull().sum()
```

```python
a.nunique()
```

```python
a.duplicated()
```

```python
X = a.iloc[:,:-1]
X
```

```python
y = a.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 25)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred =lr.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
import joblib
joblib.dump(lr,'Food_Delivery_Times.pkl')
```
